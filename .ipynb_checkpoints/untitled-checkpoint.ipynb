```python

# Define the WMS service URL and version
    
wms_url = "https://ies-ows.jrc.ec.europa.eu/effis"
version = "1.1.1"

# Create a WebMapService instance
wms = WebMapService(wms_url, version=version)
def get_image(wms, date):

    response = wms.getmap(
        layers=['ecmwf007.fwi'],
        styles=['default'],
        srs='EPSG:4326',
        bbox=(17.819824,41.590797,20.698242,43.707594),
        size=(1600, 1200),
        width=1600,
        height=1200,
        format='image/png',
        transparent=True,
        time=date,
        singletile=False,
        service='wms',
        version='1.1.1',
        request='GetMap',
        method='GET'
    )
    img_data = BytesIO(response.read())
    img = Image.open(img_data)
    return np.array(img)
    # display(plt.imshow(img))
    # img.save(f'images/{date}.png')

def get_fuels(wms):
    response = wms.getmap(
        layers=['fuel_map'],
        styles=[],
        srs='EPSG:4326',
        bbox=(17.819824,41.590797,20.698242,43.707594),
        size=(1600, 1200),
        width=1600,
        height=1200,
        format='image/png',
        transparent=True,
        time="2023-12-08/2023-12-09",
        singletile=False,
        service='wms',
        version='1.1.1',
        request='GetMap',
        method='GET'
    )
    img_data = BytesIO(response.read())
    img = Image.open(img_data)
    display(plt.imshow(img))
    # img.save(f'images/{date}.png')
```

```python
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2015, 7, 1)
end_date = date(2023, 8, 30)
for single_date in daterange(start_date, end_date):
    if single_date.strftime("%Y-%m-%d") not in dates:
        print(single_date.strftime("%Y-%m-%d"))
        try:
            images.append(get_image(wms, single_date.strftime("%Y-%m-%d")))
            dates.append(single_date.strftime("%Y-%m-%d"))
        except:
            print(f'no image {single_date.strftime("%Y-%m-%d")}')
    else:
        print(f'skipped: {single_date.strftime("%Y-%m-%d")}')
```

```python
def hexagon_corners(center,size):    
    x = center[0]
    y = center[1]

    w = math.sqrt(3) * size
    h = 2 * size

    return [ 
        (x - w / 2, y - h / 4),
        (x, y - h /2),
        (x + w/2, y - h / 4),
        (x + w/2, y + h/4),
        (x, y + h/2),
        (x - w / 2, y + h/4)
    ]

def rectangle_corners(center, w, h):
    x = center[0]
    y = center[1]

    return [
        (x - w/2, y - h/2),
        (x + w/2 , y - h/2),
        (x +w/2, y + h/2),
        (x -w /2, y + h/2)
    ]


def hexagonify(path, hexagon_size):
    im = Image.fromarray(path)
    I = np.asarray(im)
    draw = ImageDraw.Draw(im)
    
    w = math.sqrt(3) * hexagon_size
    h = 2 * hexagon_size

    #numer of hexagons horizontally and vertically
    num_hor = int(im.size[0] / w) + 2
    num_ver = int(im.size[1] / h * 4 / 3) + 2

    for i in range(0,num_hor*num_ver): 
        column = i % num_hor
        row = i // num_hor
        even = row % 2  # the even rows of hexagons has w/2 offset on the x-axis compared to odd rows.    
        
        p = hexagon_corners((column*w + even * w/2,row*h * 3/4), hexagon_size)

        # compute the average color of the hexagon, use a rectangle approximation.
        raw = rectangle_corners((column*w + even * w/2, row*h * 3/4), w, h)
        r = []
        for points in raw:
            np0 = int(np.clip(points[0], 0, im.size[0]))
            np1 = int(np.clip(points[1], 0, im.size[1]))
            r.append((np0,np1))    
        
        color = np.average(I[r[0][1]:r[3][1],r[0][0]:r[1][0]], axis=(0,1))    
        color = tuple(color.astype(np.uint8))    

        draw.polygon(p, fill=color)
    numpy_array = np.array(im)
    numpy_array = numpy_array.astype(np.uint8)
    return numpy_array


def remove_border(input_img):
    mean_image_copy = input_img.copy()
    transparent_mask = (mean_image_copy[:, :, 2] != 0)
    
    # Identify the coordinates of transparent pixels
    transparent_pixels = np.column_stack(np.where(transparent_mask))
    
    # Make four neighboring pixels transparent for each transparent pixel
    for x, y in transparent_pixels:
        neighbors = [
            (x - 1, y),
            (x + 1, y),
            (x, y - 1),
            (x, y + 1),
            (x - 2, y),
            (x + 2, y),
            (x, y - 2),
            (x, y + 2),
            (x - 3, y),
            (x + 3, y),
            (x, y - 3),
            (x, y + 3),
            (x - 4, y),
            (x + 4, y),
            (x, y - 4),
            (x, y + 4),
            (x - 5, y),
            (x + 5, y),
            (x, y - 5),
            (x, y + 5)
        ]
    
        for nx, ny in neighbors:
            if 0 <= nx < mean_image_copy.shape[0] and 0 <= ny < mean_image_copy.shape[1]:
                mean_image_copy[nx, ny, 2] = 0
    return mean_image_copy

def remove_border_2(input_img):
    mean_image_copy = input_img.copy()
    transparent_mask = (mean_image_copy[:, :, 3] == 0)
    
    # Identify the coordinates of transparent pixels
    transparent_pixels = np.column_stack(np.where(transparent_mask))
    
    # Make four neighboring pixels transparent for each transparent pixel
    for x, y in transparent_pixels:
        neighbors = [
            (x - 1, y),
            (x + 1, y),
            (x, y - 1),
            (x, y + 1),
            (x - 2, y),
            (x + 2, y),
            (x, y - 2),
            (x, y + 2),
            (x - 3, y),
            (x + 3, y),
            (x, y - 3),
            (x, y + 3),
            (x - 4, y),
            (x + 4, y),
            (x, y - 4),
            (x, y + 4),
            (x - 5, y),
            (x + 5, y),
            (x, y - 5),
            (x, y + 5)
        ]
    
        for nx, ny in neighbors:
            if 0 <= nx < mean_image_copy.shape[0] and 0 <= ny < mean_image_copy.shape[1]:
                mean_image_copy[nx, ny, 3] = 0
    return mean_image_copy
```

```python
colors = []



min_longitude = 17.819824
max_longitude = 20.698242
min_latitude = 41.590797
max_latitude = 43.707594
pixel_resolution_lon = 1600
pixel_resolution_lat = 1200
for i in range(len(hexagons['geometry'])):
    polygon = hexagons['geometry'].iloc[i]
    # Get the centroid of the polygon
    centroid = polygon.centroid
    # Get the geographic coordinates of the centroid
    centroid_lon, centroid_lat = centroid.x, centroid.y
    pixel_x = int((centroid_lon - min_longitude) / ((max_longitude - min_longitude)/pixel_resolution_lon))
    pixel_y = int((max_latitude - centroid_lat) / ((max_latitude - min_latitude)/pixel_resolution_lat))

    # Extract color from the image at the calculated pixel coordinates
    color_at_centroid = uniform_rgb_image[pixel_y, pixel_x]
    colors.append(color_at_centroid)

# Print or use the color information
hexagons['Color'] = colors

```
